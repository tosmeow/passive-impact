# Aggressive Market Impact Experiment

This notebook analyzes aggressive market impact under the hybrid propagator model:

$$P_t = P_0 + \bar{\kappa}\int_0^t G(t-s)\,d(N^a-N^b)_s + \int_0^t(\kappa(q^a_s)-\bar{\kappa})\,dN^a_s - \int_0^t(\kappa(q^b_s)-\bar{\kappa})\,dN^b_s$$

where $G$ is the Hawkes-derived propagator kernel, $\bar{\kappa}$ is constant, and $\kappa(q)$ is the instantaneous queue correction.

The market impact of adding aggressive meta orders $N^o$ is:

$$MI_t = \bar{\kappa}\int_0^t G(t-s)\,dN^o_s + \int_0^t \left[\kappa(\bar{q}_s) - \kappa(q_s)\right] \, dN_s$$
