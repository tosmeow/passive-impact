# Aggressive Market Impact Experiment

This notebook analyzes the aggressive market impact under the propagator price model:

$$P_t = \int_0^t \kappa(q_s) \, \xi(t-s) \, dN_s$$

where $\xi = \delta_0 + \psi$ is the propagator (Id + sum of exponentials) and $\kappa(q) = cq + d$.

The market impact of adding aggressive meta orders $N^o$ is:

$$MI_t = \int_0^t \left[\kappa(\bar{q}_s) - \kappa(q_s)\right] \xi(t-s) \, dN_s + \int_0^t \kappa(\bar{q}_s) \, \xi(t-s) \, dN^o_s$$
